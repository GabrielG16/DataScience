# Modificando dados

```python
#Adicionando uma nova coluna
#df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Speed']
#print(type(df))
df['Total'] =  df.iloc[:,4:10].sum(axis=1)
#df.head(5)

#Excluindo uma coluna
#df = df.drop(columns=['Total'])

#Reordenamento de colunas
cols = list(df.columns.values)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
df.head(5)
```

# Filtragem com base em mÃºltiplas condiÃ§Ãµes

```python
df.loc[(df['Type 1'] == 'Fire') & (df['Legendary'] == True)] # Condicionais entre parenteses separadamente
df.loc[(df['Type 1'] == 'Fire') | (df['Type 2'] == 'Flying')] #| = OU
new_df =  df.loc[(df['Type 1'] == 'Fire') & (df['Legendary'] == True)] #No entanto, os indices permanecem com a numeraÃ§Ã£o da tabela original
new_df.reset_index(drop=True, inplace=True) #reseta os indices ordena-los novamente na nova planilha filtrada. 
#Drop = True retira os indices anteriores que ficariam como uma nova coluna e o inplace = True realiza a mudanÃ§a no doc original
df.loc[df['Name'].str.contains('Mega')] # Filtrar elementos que contenham na coluna nome a palavra 'Mega'
df.loc[~df['Name'].str.contains('Mega')] # Filtrar elementos que NÃO contenham na coluna nome a palavra 'Mega'
```

# MudanÃ§as condicionais

# Agrupamentos de dados

# Trabalhando com grande quantidade de dados

```python

"""
"""
new_df = pd.DataFrame(columns = df.columns) # cria uma nova planilha vazia, setando as mesmas colunas do df ja criado
#Ã possÃ­vel trazer apenas pedaÃ§os dos dados para trabalhar por vez
for df in pd.read_csv('Pokemon.csv', chunksize=5):
    #print("#####\t CHUNK DF \t #####")
    #print(df)
    results = df.groupby(['Type 1']).count()['#']
    new_df = pd.concat([new_df, results])
print(new_df)

```
