```python
from nba_api.stats.static import teams
nba_teams = teams.get_teams()
nba_teams
```

```python
import pandas as pd
chaves = nba_teams[0].keys()
dicio_fin = {chave:[] for chave in chaves}
for dicio in nba_teams:
    for chave, valor in dicio.items():
        dicio_fin[chave].append(valor)
#print(dicio_fin)
df_nba = pd.DataFrame(dicio_fin)
df_nba.head(10)
# Warriors
warriors = df_nba[df_nba['nickname']== 'Warriors']
id_warriors = warriors['id'].values[0]
```

```python
dicio = {chave:[] for chave in nba_teams[0]}
for time in nba_teams:
    for criterio, valor in time.items():
        dicio[criterio].append(valor)
df = pd.DataFrame(dicio)
df.head()
```

```python
from nba_api.stats.endpoints import leaguegamefinder
gamefinder = leaguegamefinder.LeagueGameFinder(team_id_nullable=id_warriors)
games = gamefinder.get_data_frames()[0]
games.head()
```
